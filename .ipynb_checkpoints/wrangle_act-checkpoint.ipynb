## Data Gathering
In the cell below, gather **all** three pieces of data for this project and load them in the notebook. **Note:** the methods required to gather each data are different.
1. Directly download the WeRateDogs Twitter archive data (twitter_archive_enhanced.csv)

```python
import pandas as pd
import numpy as np
twitter = pd.read_csv('twitter-archive-enhanced.csv')
```

2. Use the Requests library to download the tweet image prediction (image_predictions.tsv)

```python
import requests as req
res = req.get('https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv')
with open('image_predictions.tsv', 'wb') as file:
    file.write(res.content)
images = pd.read_csv('image_predictions.tsv', sep='\t')
```

3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

tweet_ids = twitter.tweet_id.values
len(tweet_ids)

count = 0
fails_dict = {}
start = timer()

with open('tweet_json.txt', 'w') as outfile:

    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
tweets_data = []
with open('tweet-json.txt') as tweet_json_file:
    for i, line in enumerate(tweet_json_file):
        tweet_json = json.loads(line)
        data = {
            'tweet_id': int(tweet_json['id']),
            'retweet_count': int(tweet_json['retweet_count']),
            'favorite_count': int(tweet_json['favorite_count'])
        }
        tweets_data.append(data)
tweets = pd.DataFrame(tweets_data, columns=['tweet_id', 'retweet_count', 'favorite_count'])
```

## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
twitter_clean = twitter.copy()
images_clean = images.copy()
tweets_clean = tweets.copy()
```

### Issue #1:

#### Define:
in twitter dataframe: just keep rows with `retweeted_status_user_id` equals `null`, as we are more interested in OG tweets than the retweets

#### Code

```python
twitter_clean = twitter_clean[twitter_clean.retweeted_status_user_id.isna()]
```

#### Test

```python
twitter_clean
```

### Issue #2:

#### Define
only keep rows with rating denominator that equals to 10

#### Code

```python
twitter_clean = twitter_clean[twitter_clean['rating_denominator'] == 10]
```

#### Test

```python
twitter_clean.rating_denominator.value_counts()
```

### Issue #3

#### Define
drop columns `retweeted_status_id`, `retweeted_status_user_id` & `retweeted_status_timestamp` because we are not concerned in retweets

#### Code

```python
twitter_clean.drop(columns=['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], inplace=True)
```

#### Test

```python
twitter_clean.head()
```

### Issue #4

#### Define
extract the source text from href attr of the anchor tag and assigning it to the source column

#### Code

```python
def source(x):
    return x[x.find('>') + 1:][:x[x.find('>') + 1:].find('<')]

twitter_clean.source = twitter_clean.source.apply(source)
```

#### Test

```python
twitter_clean.source
```

### Issue #5

#### Define 
replace values like NaN, the, a & an with `None` in the `name` column of the twitter dataset

#### Code

```python
twitter_clean['name'] = twitter_clean['name'].replace(['an', 'NaN', 'a', 'the'], None)
```

#### Test

```python
twitter_clean.name.value_counts()
```

### Issue #6

#### Define
set the datatype of `timestamp` column in `twitter` dataset to datetime

#### Code

```python

twitter_clean['timestamp'] = pd.to_datetime(twitter_clean['timestamp'])
```

#### Test

```python
twitter_clean.info()
```

### Issue #7

#### Define
convert None value in `doggo, floofer, pupper, and puppo` to np.nan

#### Code

```python
twitter_clean['doggo'] = twitter_clean['doggo'].replace({'None': np.nan})
twitter_clean['floofer'] = twitter_clean['floofer'].replace({'None': np.nan})
twitter_clean['pupper'] = twitter_clean['pupper'].replace({'None': np.nan})
twitter_clean['puppo'] = twitter_clean['puppo'].replace({'None': np.nan})
```

#### Test

```python
twitter_clean.info()
```

### Issue #8

#### Define
dropping jpg_url duplicates in the images dataframe 

#### Code

```python
images_clean.drop_duplicates(subset=['jpg_url'], inplace=True)
```

#### Test

```python
sum(images_clean['jpg_url'].duplicated())
```

## Tidiness

### Issue #1

#### Define
melt the doggo, floofer, pupper, and puppo columns into a single column

#### Code

```python
twitter_clean.columns
twitter_clean = pd.melt(twitter_clean, id_vars=['tweet_id', 'in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp','source', 'text', 'expanded_urls', 'rating_numerator','rating_denominator', 'name'],
                                var_name='dogs_develop', value_name ='dev_stage')
twitter_clean.drop('dogs_develop', axis=1, inplace=True)
twitter_clean.drop_duplicates(subset=['tweet_id'],  keep='last')
```

#### Test

```python
twitter.info()
```

### Issue #2

### Define
join all 3 datasets/tables togather inorder to form a new dataframe.

#### Code

```python
# Perform an inner merge to get only tweets that appear in all dataframes
twitter_master =twitter_clean.merge(images_clean, how='inner', on='tweet_id').merge(tweets_clean, how='inner', on='tweet_id')
```

#### Test

```python
twitter_master['dev_stage']
```

3. Which canine received the most and least average favorite counts?

```python
import matplotlib.pyplot as plt
labels = twitter_master.dev_stage.unique().tolist()[1:]
sizes = twitter_master.groupby(['dev_stage']).favorite_count.sum()
plt.axis('equal')
plt.pie(sizes)
plt.legend(labels)
plt.title('development stages')
plt.show()
```

__we can conclude from the pie chart above that the most liked one among the dog development stages is the pupper development stage__
